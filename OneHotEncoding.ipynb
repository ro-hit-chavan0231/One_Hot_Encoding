- One Hot Encoding is the technique used to convert categorical variables into numerical format by creating binary for each veriables.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
```

## OneHotEncoding using Pandas

```python
pd.get_dummies(df,columns = ['fuel','owner'],dtype = int)
```

```python
pd.get_dummies(df,columns = ["fuel", "owner"],dtype = int, drop_first = True)
```

### OneHotEncoding with Top Categories

```python
counts = df.brand.value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts > threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, "uncommon"),dtype = int).sample(5)
```
